# Insights

## Insight 1<br>
The breeds of dogs with the highest count of tweets are:

```python
df_tweet_count = df_dogs[df_dogs["p1_dog"] == True]["p1"].value_counts().to_frame().reset_index()
df_tweet_count.columns = ["p1","tweet_count"]
df_tweet_count.nlargest(10, "tweet_count")
```

```python
golden_percentage = df_tweet_count[df_tweet_count["p1"] == "Golden_Retriever"]["tweet_count"]/df_dogs["tweet_id"].count()
golden_percentage
```

## Insight 2 <br>
The breeds of dogs with the highest count of likes and favs are:


```python
##this block merges the info from dogs and tweets in a single table, using tweet_id as key. Only the relevant rows are kept: breed and retweet+favs
df_retweets_favs_clean.rename(columns = {"id":"tweet_id"}, inplace = True)
df_retfav = pd.merge(df_dogs, df_retweets_favs_clean, how = "left",on = "tweet_id")
df_retfav = df_retfav.groupby(by = "p1").sum()["favorite_count"]+df_retfav.groupby(by = "p1").sum()["retweet_count"]
df_retfav.sort_values(ascending = False)
df_retfav = df_retfav.to_frame().reset_index()
df_retfav.columns = ["p1","retweet_and_favs"]
df_retfav
```

To get a realistic idea of how popular each breed of dog is, we need to normalize this value by the number of posts of each breed.

```python
##this block calculates likeability as the average retweets and fav per post.
df_popular = pd.merge(df_retfav, df_tweet_count, how = "left", on = "p1")
df_popular["likeability"] = df_popular["retweet_and_favs"]/df_popular["tweet_count"]
df_popular.nlargest(10,"likeability")
```

## Insight 3 <br>
The most popular dog name in the dataset is:

```python
df_popular_names = df_dogs["name"].value_counts().to_frame().reset_index()
df_popular_names.columns = ["name","name_count"]
df_popular_names = df_popular_names[df_popular_names["name"]!= "None"]
df_popular_names.nlargest(10,"name_count")
```
